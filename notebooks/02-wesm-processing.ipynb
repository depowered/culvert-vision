```python
from pathlib import Path
import geopandas
```

```python
data_dir = Path("../data/")
source = data_dir / "external" / "usgs" / "WESM.gpkg.zip"
output = data_dir / "interim" / "mn_wesm.gpkg"
```

```python
filter = "workunit LIKE 'MN%' AND (ql = 'QL 1' OR ql = 'QL 0')"
gdf: geopandas.GeoDataFrame = geopandas.read_file(filename=source, where=filter)
```

```python
gdf.to_file(output, driver="GPKG")
```
